```python
def shuffle(input, offset):
    right = min(len(input)-1,offset + 3) - offset
    left = 3 - right
    #print("left/right",left, right)
    three = input[offset+1 : offset+right+1]
    three.extend(input[0:left])    
    rest = input[:]
    for c in three:
        rest.remove(c)
    #print("rest/three", rest, three)
    destinationCup = input[offset] - 1
    if destinationCup < 1:
        destinationCup = 9
    while not destinationCup in rest:
        destinationCup -= 1
        if destinationCup < 1:
            destinationCup = 9
    nextCupIndex = (offset + 4) % len(input)
    nextCupValue = input[nextCupIndex]
    #print("target cup is", destinationCup)
    #print("next cup is", nextCupValue, "(at old index", nextCupIndex, ")")
    result = []    
    for i in range(0,len(rest)):
        if rest[i] == destinationCup:
            result = rest[0:i+1]
            result.extend(three)
            result.extend(rest[i+1:len(rest)])
            break
    newOffset = 0
    for i in range(0,len(result)):
        if result[i] == nextCupValue:
            newOffset = i
            break
    #print(result, newOffset)
    expectedOffset = (offset + 1) % len(input)
    diff = newOffset - expectedOffset
    while (diff < 0):    
        diff += 9
    movedResult = result[diff:len(result)]
    movedResult.extend(result[0:diff])
    return (movedResult, expectedOffset)
```

```python
def shuffle_n(input, n):
    data = input[:]
    current = 0
    for i in range(0,n):
        next = shuffle(data, current)
        #print(next)
        data = next[0]
        current = next[1]        
        #print("")
        #printCups(data, current)
        #print("")
    return (data, current)
```

```python
def smart_pop(data):
    # no need for empty lists at start
    while (len(data[0]) == 0):
        data.pop(0)
    # get the first one...
    return data[0].pop(0)

def smart_push(data, value, n):
    if (len(data[-1]) < math.sqrt(n)):
        data[-1].append(value)
    else:
        data.append([value])

def smart_insert(data, three, destinationCup, n):
    for i in range(0,len(data)):
        chunk = data[i]
        for j in range(0,len(chunk)):            
            if chunk[j] == destinationCup:
                newChunk = chunk[0:j+1]
                newChunk.extend(three)
                newChunk.extend(chunk[j+1:len(chunk)])
                data[i] = newChunk
                return

def shuffle_part2(data, n):
    #print("Shuffling",data,n)
    first = smart_pop(data)
    three = []
    for i in range(0,3):
        three.append(smart_pop(data))
    #print(first, three, data)
    smart_push(data, first, n)    
    #print(three, data)

    destinationCup = first - 1
    if destinationCup < 1:
        destinationCup = n
    while destinationCup in three:
        destinationCup -= 1
        if destinationCup < 1:
            destinationCup = n

    #print("destination cup",destinationCup)
    smart_insert(data, three, destinationCup,n)

def shuffle_part2_multiple(data, n, total):
    print("Shuffle multiple", n, total)
    for i in range(0,total):
        shuffle_part2(data, n)
        if i % 100 == 0:
            print ("iteration" , i, i*100.0/total)
        # print(data)
    return data
```

```python
def generateStartingDeck(prefix, n, chunk_size):
    fullList = prefix[:]    
    fullList.extend(range(len(prefix)+1, n+1))
    result = []
    while len(fullList) > 0:
        result.append(fullList[0:chunk_size])
        fullList = fullList[chunk_size:]
    return result
```

```python
def generateNeighborList(input, n):
    result = {}
    for i in range(0,len(input)):
        value = input[i]
        left = None
        if i == 0:
            if n > len(input):
                left = n
            else:
                left = input[n-1]
        else:
            left = input[i-1]
        right = None
        if i == len(input) - 1:
            if n > len(input):
                right = len(input) + 1
            else:
                right = input[0]
        else:
            right = input[i+1]

        result[value] = (left, right)
    
    if n > len(input):
        result[len(input)+1] = (input[-1], len(input)+2)
        for i in range(len(input)+2, n):
            result[i] = (i-1,i+1)
        result[n] = (n-1, input[0])

    return result
```

```python

def neighbor_pop(data, item):
    #print("POP", data, item)
    neighbors = data[item]
    left = neighbors[0]
    right = neighbors[1]
    oldLeftNeighbors = data[left]
    oldRightNeighbors = data[right]
    data[left] = (oldLeftNeighbors[0], right)
    data[right] = (left, oldRightNeighbors[1])
    data[item] = None
    return (item, right)

def neighbor_insert(data, left, item):
    oldLeftNeighbors = data[left]
    right = oldLeftNeighbors[1]
    oldRightNeighbors = data[oldLeftNeighbors[1]]
    data[left] = (oldLeftNeighbors[0], item)
    data[right] = (item, oldRightNeighbors[1])
    data[item] = (left, right)

def advance_neighbors(data, current, n):
    first = current
    next = data[current][1]    
    three = []
    for i in range(0,3):
        temp = neighbor_pop(data, next)
        three.append(temp[0])
        next = temp[1]
    
    #print("FIRST/THREE/NEXT", first, three, next, data)

    destinationCup = first - 1
    if destinationCup < 1:
        destinationCup = n
    while destinationCup in three:
        destinationCup -= 1
        if destinationCup < 1:
            destinationCup = n
    
    neighbor_insert(data, destinationCup, three[0])
    neighbor_insert(data, three[0], three[1])
    neighbor_insert(data, three[1], three[2])

    return next

def advance_neighbors_multiple(data, start, total):
    n = len(data)
    print("Advance multiple", n, total)
    current = start
    for i in range(0,total):
        next = advance_neighbors(data, current, n)
        #print(data)
        current = next
        if i % 100000 == 0:
            print ("iteration" , i, i*100.0/total)
    star1 = data[1][1]
    star2 = data[star1][1]
    print("STARS",star1, "*", star2, "=", star1*star2)
```
