```python
def parseRule(line):
    parts = line.split(": ")
    key = int(parts[0])
    value = parts[1]
    if value.count('"') > 0:
        return (key,value.strip('" '))
    value_parts = value.split("|")
    subrules = []
    for v in value_parts:        
        temp = list(map(int,v.strip().split(" ")))
        subrules.append(temp)
    return (key, subrules)    

def parse(input):
    rules = {}
    data = []
    for line in input:
        if line.count(":") > 0:
            item = parseRule(line)
            rules[item[0]] = item[1]
        elif len(line) > 0:
            data.append(line)
    return (rules, data)
```

```python
def matchesRulesChoice2(rules, text, choice):
    if len(text) < 2:
        return False
    if len(choice) != 2:
        throw
    for i in range(1,len(text)):
        left = text[:i]
        right = text[i:]
        if len(left) < len(right):
            # first check the small part
            if not matchesRules(rules, left, choice[0]):
                continue
            if not matchesRules(rules, right, choice[1]):
                continue
        else:
            if not matchesRules(rules, right, choice[1]):
                continue
            if not matchesRules(rules, left, choice[0]):
                continue            
        #print("MATCH",text, choice)
        return True
    #print("FAIL",text, choice)
    return False

def matchesRulesChoice3(rules, text, choice):    
    if len(text) < 3:
        return False
    if len(choice) != 3:
        throw
    for i in range(1,len(text)-1):
        for j in range(i+1,len(text)):
            left = text[:i]
            middle = text[i:j]
            right = text[j:]
            #print("three parts",left,middle, right)
            if not matchesRules(rules, left, choice[0]):
                continue
            if not matchesRules(rules, middle, choice[1]):
                continue
            if not matchesRules(rules, right, choice[2]):
                continue
            #print("MATCH",text, choice)
            return True
    #print("FAIL",text, choice)
    return False

def matchesRulesChoice(rules, text, choice):
    if len(choice) == 1:
        return matchesRules(rules, text, choice[0])
    elif len(choice) == 2:
        return matchesRulesChoice2(rules, text, choice)
    elif len(choice) == 3:
        return matchesRulesChoice3(rules, text, choice)
    return False

def matchesRules(rules, text, r):
    rule = rules[r]
    if (isinstance(rule, list)):
        # recurse        
        for choice in rule:
            if matchesRulesChoice(rules, text, choice):
                return True
        return False
    else:
        return text == rule
```

```python
def solvePartOne(input):
    (rules, data) = parse(input)
    count = 0
    for text in data:
        match = matchesRules(rules,text,0)
        print(text, match)
        if match:
            count += 1
    return count
```
